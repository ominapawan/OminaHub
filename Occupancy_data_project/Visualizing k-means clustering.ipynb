```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

import matplotlib as mpl
import matplotlib.style
mpl.style.use('seaborn-bright')
%matplotlib inline

plt.rcParams['figure.figsize'] = (8, 6)
```

```python

# Creating a sample dataset with 3 clusters
X, y = make_blobs(n_samples=400, n_features=3, centers= [(22, 28, 700), (12, 20, 10), (18, 25, 400)], cluster_std=4)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
ax.set_yticks([12, 16, 20, 24, 28, 32])
ax.set_yticklabels([12, 16, 20, 24, 28, 32])
ax.view_init(30, 30)
ax.set_xlabel('Temperature')
ax.set_ylabel('Humidity')
ax.set_zlabel('Light')
plt.savefig("test_kmean1.svg", format="svg")
!gsutil cp 'test_kmean1.svg' 'gs://omina-test-set/occupancy-data/fig_kmean1.svg'
```

```python
# Initializing KMeans
kmeans = KMeans(n_clusters=3)
# Fitting with inputs
kmeans = kmeans.fit(X)
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap = 'winter')
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c='red', s=1000)
ax.set_yticks([12, 16, 20, 24, 28, 32])
ax.set_yticklabels([12, 16, 20, 24, 28, 32])
ax.view_init(30, 30)
ax.set_xlabel('Temperature')
ax.set_ylabel('Humidity')
ax.set_zlabel('Light')
plt.savefig("test_kmean2.svg", format="svg")
!gsutil cp 'test_kmean2.svg' 'gs://omina-test-set/occupancy-data/fig_kmean2.svg'
```
